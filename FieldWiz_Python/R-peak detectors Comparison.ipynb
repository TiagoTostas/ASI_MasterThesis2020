## Plot R-Peaks detected with different methods

```python
# Run through all the files
directory = r"/Users/tiagorodrigues/OneDrive - Universidade de Lisboa/TiagoRodrigues_EPFL_FieldWiz_tese_2020/Data/Annotations_Joana/Annotated"
fs = 250
time_window = 0.1 
samples_window = np.round(fs * time_window) 

# Recording: 20200505-TR-Belt-Annotated.hdf5

file = os.listdir(directory)[1];
filename = os.path.join(directory, file)

f = h5py.File(filename,'r')

# Get ECG and timestamps from hdf5
ecg = f['signals/ECG/raw/ecg'][:]
    
# reverse polarity for engzee to work (ECG has inverted polarities)
ecg = ecg * (-1)

ti = 1000;
tf = 1010;
fs = 250;
ecg = ecg [ti * fs : tf *fs]

plt.subplots(figsize=(10,10));
plt.plot(ecg,linewidth = 3)
plt.xlabel('Samples',fontsize=16)
plt.ylabel('ADC Value',fontsize=16)

methods = ['PanTompkins','Christov','Gamboa','Engzee','Elgendi','Kalidas']
colors = ['green','blue','red','black','purple','orange']


for i in range(6):
        
    # Get annotations from the detector
    # Pth = 5 and N = 5, only used for rodrigues2020        
    detector_peaks = nk.ecg_findpeaks(5, 5, ecg, sampling_rate = fs, method = methods[i], show = False)
        
    randcolor = colors[i];
    
    t = True
    for xc in detector_peaks['ECG_R_Peaks']:
        if t:
            t = False
            plt.axvline(x=xc, color = randcolor,linestyle = '-.',label = methods[i])
        else:
            plt.axvline(x=xc, color = randcolor,linestyle = '-.')
    
f.close()


plt.legend( loc ="upper right",prop={'size': 16})
plt.xlim((2000, 2250)) 
```
